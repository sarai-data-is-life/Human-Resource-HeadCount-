```python
def expand_area(Area):
    if Area == 'CORP':
        return 'Corporate' 
    elif Area == "NCSA":
        return 'North, Central and South America'
    elif Area == "APAC":
        return "Asia Pacific"
    elif Area == "MAO":
        return "Marine"
    elif Area == "MENA":
        return "Middle East and North Africa"
    elif Area == "EARC":
        return "Europe, Africa, Russia and Caspian"
    elif Area == "TECH":
        return "Technology"
    elif Area == "LUTECH":
        return "Lutech"
    else:
        return 'Area not Provided'
    
```

```python
area_names = []
for item in list(headcount_data['Area']):
    area_names.append(expand_area(item))
```

```python
headcount_data['Area'] = area_names
```
